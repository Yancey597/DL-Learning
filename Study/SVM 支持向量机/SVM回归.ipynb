```python
from sklearn.svm import LinearSVR, SVR

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(666)
x = np.random.uniform(-3,3,size=100)
y = 0.5 * x**2 +x +2 +np.random.normal(0,1,size=100)
X = x.reshape(-1,1)
```

```python
plt.scatter(X,y,c='b')
plt.show()
```

```python
from sklearn.pipeline import make_pipeline
pipeline = make_pipeline(
    StandardScaler(),
    LinearSVR(epsilon=0.01)
)

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=666)

pipeline.fit(x_train, y_train)
```

```python
pipeline.score(x_test, y_test)
```

```python
plt.scatter(X, y,c='blue')

y_pred = pipeline.predict(X)
plt.plot(np.sort(x), y_pred[np.argsort(x)], c='red')
```

```python
svr = SVR(kernel='rbf', C=1.0, epsilon=0.01)
pipeline1 = make_pipeline(
    StandardScaler(),
    SVR(kernel='rbf', C=1.0, epsilon=0.01)
)

pipeline1.fit(x_train, y_train)
pipeline1.score(x_test, y_test)
```

```python
plt.scatter(X, y,c='blue')
y_pred1 = pipeline1.predict(X)
plt.plot(np.sort(x), y_pred1[np.argsort(x)], c='red')
```
