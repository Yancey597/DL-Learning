```python
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
house =  fetch_california_housing()
```

```python
house
```

```python
X = house.data
y = house.target
```

```python
X
```

```python
linear_reg = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y)
```

```python
linear_reg.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt
```

```python
linear_reg.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, linear_reg.predict(x_test))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, linear_reg.predict(x_test))
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, linear_reg.predict(x_test))
```

```python
from sklearn.metrics import root_mean_squared_error
```

```python
root_mean_squared_error(y_test, linear_reg.predict(x_test))
```

## 多项式回归

```python
x = np.random.uniform(-4,2, size=100)
```

```python
y = 2*x**2 + 4*x + np.random.randn(100)
```

```python
X = x.reshape(-1,1)
```

```python
X_2 = X**2
```

```python
data = np.hstack((X, X_2))
```

```python
LinearReg = LinearRegression()
```

```python
LinearReg.fit(data, y)
```

```python
LinearReg.coef_
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
X
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
poly.fit(X)
```

```python
data = poly.transform(X)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, y, test_size=0.7, random_state=42)
```

```python
linear_reg.fit(x_train, y_train)
```

```python
linear_reg.score(x_test, y_test)
```
